## 2. Library Imports Test

## 3. Data Visualization Test

```python
# Create some sample data and plot
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', alpha=0.7, label='Noisy sine wave')
plt.plot(x, np.sin(x), 'r--', label='Clean sine wave')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sample Plot - Environment Test')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 4. Pandas DataFrame Test

```python
# Create and manipulate a sample DataFrame
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'age': [25, 30, 35, 28, 32],
    'city': ['New York', 'San Francisco', 'Chicago', 'Austin', 'Seattle'],
    'score': [85, 92, 78, 95, 88]
}

df = pd.DataFrame(data)
print("Sample DataFrame:")
print(df)
print(f"\nDataFrame shape: {df.shape}")
print(f"Average score: {df['score'].mean():.1f}")
```

## 5. API Request Test

```python
# Test making an API request
try:
    response = requests.get('https://httpbin.org/json')
    if response.status_code == 200:
        data = response.json()
        print("✅ API request successful!")
        print(f"Response keys: {list(data.keys())}")
    else:
        print(f"❌ API request failed with status: {response.status_code}")
except Exception as e:
    print(f"❌ Error making API request: {e}")
```
